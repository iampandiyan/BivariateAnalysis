# 7)Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)
# Pvalue is 1.9407478346570754e-24 > 0.05 Hence, There is a significance difference between the etest_p and mba_p of students

```python
#One Way Classification
import scipy.stats as stats
f_statistic,p_value=stats.f_oneway(dataset['etest_p'],dataset['mba_p'])
print("F-statistic:", f_statistic)
print("p-value:", p_value)
H0='There is no significance difference between the etest_p and mba_p of students'
H1='There is a significance difference between the etest_p and mba_p of students'
if p_value < 0.05:
    print(H1)
else:
    print(H0)
```

# 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make decision using Hypothesis Testing)


```python
from scipy.stats import ttest_rel
etest_p=dataset[dataset['degree_t']=='Sci&Tech']['etest_p']
mba_p=dataset[dataset['degree_t']=='Sci&Tech']['mba_p']
stat,pval=ttest_rel(etest_p,mba_p)
print(pval)
H0='There is no similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%'
H1='There is similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%'
if(pval<0.05):
    print(H1)
else:
    print(H0)
```

# 11)Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)


```python
from scipy.stats import ttest_rel
etest_p=dataset[dataset['degree_t']=='Sci&Tech']['etest_p']
mba_p=dataset[dataset['degree_t']=='Sci&Tech']['mba_p']
stat,pval=ttest_rel(etest_p,mba_p)
print(pval)
H0='There is no similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%'
H1='There is similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%'
if(pval<0.05):
    print(H1)
else:
    print(H0)
```

# 12)Which parameter is highly correlated with salary?
