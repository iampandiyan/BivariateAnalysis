```python
import pandas as pd
dataset=pd.read_csv('PreprocessedStudentsDataset.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
#H0 -> There is no significance difference between the ssc_p, hsc_p and degree_p of students
#H1 -> There is a significance difference between the ssc_p, hsc_p and degree_p of students
```

```python
#One Way Classification
import scipy.stats as stats
f_statistic,p_value=stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
print("F-statistic:", f_statistic)
print("p-value:", p_value)
H0='There is no significance difference between the ssc_p, hsc_p and degree_p of students'
H1='There is a significance difference between the ssc_p, hsc_p and degree_p of students'
if p_value < 0.05:
    print(H1)
else:
    print(H0)
```

```python
#Two Way Classification
# Convert the data into a DataFrame
import statsmodels.api as sm
from statsmodels.formula.api import ols
df = pd.DataFrame(dataset)

# Define the model and fit it
model = ols('etest_p ~ C(ssc_p) + C(hsc_p) + C(ssc_p):C(hsc_p)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-Way ANOVA results:")
print(anova_table)
```

```python
#Two Way Classification
# Convert the data into a DataFrame
import statsmodels.api as sm
from statsmodels.formula.api import ols
df = pd.DataFrame(dataset)

# Define the model and fit it
model = ols('salary ~ C(ssc_p) + C(hsc_p) + C(ssc_p):C(hsc_p)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("Two-Way ANOVA results:")
print(anova_table)
```
