```python
import pandas as pd
dataset=pd.read_csv('PreprocessedStudentsDataset.csv')
dataset
```

```python
dataset[dataset['gender']=='M']
```

```python
dataset[dataset['gender']=='F']
```

```python
#INdependent Sample Unpaired T Test
#Different group(Male,Female) but same condition(salary)
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male=dataset[dataset['gender']=='M']['salary']
female=dataset[dataset['gender']=='F']['salary']
ttest_ind(male,female)
```

```python
#dependent Sample Paired T Test
#Same Group(Male) different conditions(ssc_p,hsc_p)
from scipy.stats import ttest_rel
male_ssc_p=dataset[dataset['gender']=='M']['ssc_p']
male_hsc_p=dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male_ssc_p,male_hsc_p)
```

```python
#Assignment
#dependent Sample Paired T Test
#Same Group(Commerce) different conditions(etest_p,mba_p
from scipy.stats import ttest_rel
Commerce_etest_p=dataset[dataset['hsc_s']=='Commerce']['etest_p']
Commerce_mba_p=dataset[dataset['hsc_s']=='Commerce']['mba_p']
ttest_rel(Commerce_etest_p,Commerce_mba_p)
```
